```python
# 1) Mount en paden
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import os, shutil, glob
BASE = "/content/drive/MyDrive/MindVault-AI"
store_dir = os.path.join(BASE, "store")
os.makedirs(store_dir, exist_ok=True)

# 2) Verplaats eventuele lokaal gemaakte bestanden uit /store -> Drive/store
for p in ["/store/lottery_mindfuck.txt", "/store/lottery_pack.zip"]:
    if os.path.exists(p):
        shutil.move(p, os.path.join(store_dir, os.path.basename(p)))

# 3) Schrijf voortaan rechtstreeks in Drive/store
lottery_file = os.path.join(store_dir, "lottery_mindfuck.txt")
zip_file = os.path.join(store_dir, "lottery_pack.zip")

with open(lottery_file, "w", encoding="utf-8") as f:
    f.write("MindVault-AI Global Lottery Pack – DRIVE VERSION\n")

import zipfile
with zipfile.ZipFile(zip_file, "w") as z:
    z.write(lottery_file, arcname=os.path.basename(lottery_file))

# 4) Controleren
print("Drive store listing:")
print(os.listdir(store_dir))

# === Sexy Teaser Pack → Drive/MindVault-AI/store ===
# Audio + cover + beschrijving + zip, NL+EN. 100% lokaal in Drive.
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import os, zipfile, textwrap
BASE = "/content/drive/MyDrive/MindVault-AI"
STORE = os.path.join(BASE, "store"); os.makedirs(STORE, exist_ok=True)

# 1) Dependencies voor audio
!apt-get -y install ffmpeg >/dev/null
!pip -q install gTTS pydub python-docx Pillow

# 2) Teaser script (NL + EN, kort en zwoel)
nl_script = ("Hey jij... stop met scrollen. "
             "Zin in iets ondeugends? "
             "Luister en ontdek waarom dit jouw favoriete nacht wordt. "
             "MindVault—speel mee, win slim.")
en_script = ("Hey you... stop scrolling. "
             "In the mood for trouble? "
             "Listen and discover why tonight is yours. "
             "MindVault—play smart, win smarter.")

# 3) Maak MP3 met stem + zacht soundbed
from gtts import gTTS
from pydub import AudioSegment
from pydub.generators import Sine

def build_teaser(text, out_path, voice_lang):
    # TTS
    tts_path = out_path.replace(".mp3", "_voice.mp3")
    gTTS(text=text, lang=voice_lang).save(tts_path)
    voice = AudioSegment.from_file(tts_path)

    # Zachte pad‑tone als bed (−18 dB)
    bed = Sine(220).to_audio_segment(duration=len(voice)+500).apply_gain(-18)
    mix = bed.overlay(voice)

    # Fade in/out en export 128 kbps
    final = mix.fade_in(400).fade_out(600)
    final.export(out_path, format="mp3", bitrate="128k")
    return out_path

nl_mp3 = os.path.join(STORE, "sexy_teaser_NL.mp3")
en_mp3 = os.path.join(STORE, "sexy_teaser_EN.mp3")
build_teaser(nl_script, nl_mp3, "nl")
build_teaser(en_script, en_mp3, "en")
print("🎧 MP3 ready:", os.path.basename(nl_mp3), os.path.basename(en_mp3))

# 4) Cover visual (PNG)
from PIL import Image, ImageDraw, ImageFont
W,H = 1600, 900
img = Image.new("RGB",(W,H),(9,9,12))
d = ImageDraw.Draw(img)
# eenvoudige gradient
for y in range(H):
    c = int(12+ (y/H)*60)
    d.line([(0,y),(W,y)], fill=(c, c-2, c+10))
# tekst
title = "SEXY TEASER"
sub   = "MindVault‑AI • Low Entry • High Pull"
try:
    font1 = ImageFont.truetype("DejaVuSans-Bold.ttf", 120)
    font2 = ImageFont.truetype("DejaVuSans.ttf", 44)
except:
    font1 = ImageFont.load_default(); font2 = ImageFont.load_default()
# Calculate text size using textbbox
bbox1 = d.textbbox((0, 0), title, font=font1)
tw1, th1 = bbox1[2] - bbox1[0], bbox1[3] - bbox1[1]
d.text(((W-tw1)//2, H*0.30), title, font=font1, fill=(255,220,240))
bbox2 = d.textbbox((0, 0), sub, font=font2)
tw2, th2 = bbox2[2] - bbox2[0], bbox2[3] - bbox2[1]
d.text(((W-tw2)//2, H*0.52), sub, font=font2, fill=(240,240,255))
cover_path = os.path.join(STORE, "sexy_teaser_cover.png")
img.save(cover_path)
print("🖼️ Cover ready:", os.path.basename(cover_path))

# 5) Beschrijving (TXT + DOCX)
desc_nl = """SEXY TEASER – MindVault‑AI
Lage instap. Hoge verleiding. Download de teaser, zet je publiek ‘aan’
en stuur ze door naar je core offer. Free‑entry optie beschikbaar.
Disclaimer: geen gegarandeerde winsten; 18+; waar verboden niet geldig.
"""
desc_en = """SEXY TEASER – MindVault‑AI
Low entry. High pull. Use the teaser to hook attention and push to core offers.
Free‑entry available. Disclaimer: no guaranteed profits; 18+; void where prohibited.
"""
txt_path = os.path.join(STORE, "sexy_teaser_description.txt")
with open(txt_path,"w",encoding="utf-8") as f:
    f.write(desc_nl+"\n---\n"+desc_en)

from docx import Document
docx_path = os.path.join(STORE, "sexy_teaser_page.docx")
doc = Document(); doc.add_heading("Sexy Teaser – MindVault‑AI", 0)
doc.add_paragraph(desc_nl); doc.add_paragraph(desc_en); doc.save(docx_path)
print("📝 Copy ready:", os.path.basename(txt_path), os.path.basename(docx_path))

# 6) Zip alles
zip_path = os.path.join(STORE, "sexy_teaser_pack.zip")
with zipfile.ZipFile(zip_path,"w") as z:
    for p in [nl_mp3, en_mp3, cover_path, txt_path, docx_path]:
        z.write(p, arcname=os.path.basename(p))
print("📦 ZIP ready:", os.path.basename(zip_path))

# 7) Toon listing
print("Store listing:", os.listdir(STORE))
```

```python
# === MindVault‑AI OneCell (setup + UI + acties) ===
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import os, time, threading, datetime, zipfile
import ipywidgets as widgets
from IPython.display import display

STORE = "/content/drive/MyDrive/MindVault-AI/store"
os.makedirs(STORE, exist_ok=True)
LOCK  = os.path.join(STORE, ".running.lock")

def keep_alive():
    while os.path.exists(LOCK):
        time.sleep(60)
        print("⏱ keep-alive", datetime.datetime.now().strftime("%H:%M:%S"))

def safe_run(task):
    if os.path.exists(LOCK):
        print("⛔ Al bezig. Wacht tot run klaar is.")
        return
    open(LOCK,"w").close()
    threading.Thread(target=keep_alive, daemon=True).start()
    try:
        task()
    finally:
        if os.path.exists(LOCK):
            os.remove(LOCK)
        print("✅ Run afgerond")

def make_hello():
    p = os.path.join(STORE, "hello.txt")
    with open(p, "a", encoding="utf-8") as f:
        f.write(f"✅ MindVault‑AI testbestand @ {datetime.datetime.now().isoformat()}\n")
    print("Testbestand:", p)

def build_index_and_zip():
    idx = os.path.join(STORE, "index.csv")
    with open(idx, "w", encoding="utf-8") as f:
        f.write("Path,Size,Modified\n")
        for root, _, files in os.walk(STORE):
            for name in files:
                fp = os.path.join(root, name)
                st = os.stat(fp)
                f.write(f"{fp},{st.st_size},{datetime.datetime.fromtimestamp(st.st_mtime).isoformat()}\n")
    zip_name = os.path.join(STORE, f"MindVaultAI_All_{datetime.datetime.now():%Y%m%d_%H%M}.zip")
    with zipfile.ZipFile(zip_name, "w") as z:
        for root, _, files in os.walk(STORE):
            for name in files:
                fp = os.path.join(root, name)
                z.write(fp, arcname=os.path.relpath(fp, STORE))
    print("Index en ZIP klaar:", idx, "—", zip_name)

btn_zip = widgets.Button(description="🚀 Run MindVault‑AI Generator", button_style="success")
btn_zip.on_click(lambda b: safe_run(build_index_and_zip))
display(btn_zip)

btn_test = widgets.Button(description="🧪 Testbestand maken")
btn_test.on_click(lambda b: safe_run(make_hello))
display(btn_test)

print("STORE:", STORE)
```
